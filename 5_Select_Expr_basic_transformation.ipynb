```python
from pyspark.sql.types import *

schema_string ="name string, age int"

schema_spark = StructType([
    StructField("name",StringType(),True),
    StructField("age",IntegerType(),True)
])


```

```python
#Column and Expression

from pyspark.sql.functions import col , expr

emp["salary"]
```

```python
# Select columns 
emp_filtered= emp.select(col("employee_id"),expr("name"),emp.age,emp.salary)\

emp_filtered.show()
```

```python
emp_filtered.printSchema()
```

```python
emp_cast = emp_filtered.select(expr("employee_id as emp_id"),emp.name,expr("cast(age as int) as age"),col("salary"))
```

```python
emp_cast.show()
emp_cast.printSchema()
```

```python
emp_casted_1 = emp_filtered.selectExpr("employee_id as emp_id","name","cast(age as int) as age","salary")

emp_casted_1.show()

emp_casted_1.printSchema()
```

```python
emp_final = emp_filtered.select("employee_id","name","age","salary").where("age >30")
```
